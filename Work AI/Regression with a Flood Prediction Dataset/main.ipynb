```python
def normalization(df):
    df_min_max_scaled = df.copy() 
  
    for column in df_min_max_scaled.columns: 
        df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())     
    
    return df_min_max_scaled;
```

```python
df = pd.read_csv("train.csv");
# df = normalization(df);
display(df);
```

```python
train, test = train_test_split(df, test_size=0.3);

train_f = train.drop(columns=['FloodProbability', 'id']);
train_l = train['FloodProbability'];

test_f = test.drop(columns=['FloodProbability', 'id']);
test_l = test['FloodProbability'];
```

```python
model = linear_model.LinearRegression().fit(train_f, train_l);
```

```python
print(model.score(test_f, test_l));
```

```python
input_r = pd.read_csv("test.csv");

ans = pd.DataFrame(model.predict(input_r.drop(columns=['id'])));
ans.to_csv("ans.csv", index=False);
```
