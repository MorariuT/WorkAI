```python
import matplotlib.pylab as plt
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()


data["Target"] = enc.fit_transform(data["Target"]);

display(data);

plt.matshow(data.corr());
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler();
data = data.astype(float);

norm_data = data.drop(columns=["Target"]);

norm_data = pd.DataFrame(scaler.fit_transform(norm_data.values), columns=norm_data.columns, index=norm_data.index)

data = data["Target"];

display(norm_data);
display(data);

norm_data["Target"] = data;

display(norm_data)
#data = norm_data.merge(data, right_on=True);

#data = (data - data.min()) / (data.max() - data.min())
```

```python
from sklearn.neural_network import MLPClassifier;


X = norm_data.drop(columns=["Target"]).values.tolist();
y = norm_data["Target"].values.tolist();

#print(X);
print(y);

model = MLPClassifier(verbose=1, tol=1e-5, learning_rate_init=0.0001, learning_rate="adaptive", max_iter=10000, batch_size=64);
model.fit(X, y);
```

```python
def GraduateApply(row):
    if(row["Target"] == 0):
        row["New_Target"] = "Dropout";
    elif(row["Target"] == 1):
        row["New_Target"] = "Enrolled";
    else:
        row["New_Target"] = "Graduate";
    return row;
```

```python
data = pd.read_csv("test.csv");

ids = data["id"];

data.drop(columns=["id"], inplace=True);
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler();
data = data.astype(float);


data = pd.DataFrame(scaler.fit_transform(data.values), columns=data.columns, index=data.index)



res = model.predict(data);







ans = pd.DataFrame(res);

ans["id"] =  ids;
ans["Target"] = ans[0];
ans = ans.apply(GraduateApply, axis=1);

ans.drop(columns=[0, "Target"], inplace=True);



display(ans);

ans.to_csv("ans.csv", index=False);
```
