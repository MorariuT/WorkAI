```python
def processRow(row):
    global iterr;

    if(iterr % 1500 == 0):
        print((iterr) * 100 / 200000, "%  processed");
    date = row['pickup_datetime']

    hour = str(date).split(':')[0].split(' ')[1];
    row['hour'] = hour;

    start_long = float(row['pickup_longitude'])
    start_latd = float(row['pickup_latitude'])

    end_long = float(row['dropoff_longitude'])
    end_latd = float(row['dropoff_latitude'])

    if(start_latd < -90 or start_latd > 90 or start_long < -90 or start_long > 90
    or end_latd < -90 or end_latd > 90 or end_long < -90 or end_long > 90):
        return None;


    if(start_latd > 25 and start_latd < 75 and start_long > -150 and start_long < -25): row['avrage income'] = 4000;
    elif(start_latd > 25 and start_latd < 75 and start_long > -25 and start_long < -50): row['avrage income'] = 2200;
    elif(start_latd > -25 and start_latd < 25 and start_long > -100 and start_long < -50): row['avrage income'] = 1200;
    elif(start_latd > -25 and start_latd < 25 and start_long > -25 and start_long < 25): row['avrage income'] = 700;
    else: row['avrage income'] = 1000;
    

    #print(start_latd, start_long)

    row["dist"] = int(geopy.distance.geodesic((start_long, start_latd), (end_long, end_latd)).km)
    
    # location = geolocator.reverse(str(start_latd) + "," + str(start_long))
    # row['avrage income'] = 200
    # for obj in average_monthly_incomes:
    #     if(obj['country'] in location.address):
    #         row['avrage income'] = obj['income'];
    #         break;

    
    
    #print(location.address)
    iterr += 1;

    return row;

def selectNewYork(row):
    min_longitude, max_longitude = -74.05, -73.75
    min_latitude, max_latitude = 40.63, 40.85

    if(row['pickup_latitude'] > min_latitude and row['pickup_latitude'] < max_latitude 
    and row['pickup_longitude'] > min_longitude and row['pickup_longitude'] < max_longitude): return row;


def normalization(df):
    df_min_max_scaled = df.copy() 
  
    for column in df_min_max_scaled.columns: 
        df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())     
    
    return df_min_max_scaled;
        
```

```python
from sklearn.model_selection import train_test_split


train, test = train_test_split(data, test_size=0.3);


labels_train = train['fare_amount'];
labels_test = test['fare_amount'];

train.drop(columns=['fare_amount'], inplace=True);

train = normalization(train);

display(train);

test.drop(columns=['fare_amount'], inplace=True);
test  = normalization(test);
display(test);
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(train, labels_train);
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print(model.predict(test.head(4)));
print(labels_test.head(4));

print(r2_score(labels_test, model.predict(test)));
print(mean_squared_error(labels_test, model.predict(test)));
```
