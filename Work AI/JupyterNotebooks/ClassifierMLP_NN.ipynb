```python
from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import random
from mpl_toolkits.mplot3d import Axes3D
from sklearn import preprocessing
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
def normalize(arr, max):
    for i in range(len(arr)):
        arr[i][0] = arr[i][0] / max;
        arr[i][1] = arr[i][1] / max;
```

```python
points = [];
labels = [];
roots = [[0, 0], [1000, 1000], [0, 1000], [1000, 0]];
traing_data = []


it = 1;
for point in roots:
    for i in range(300):
        diff_x = random.randint(-400, 400);
        diff_y = random.randint(-400, 400);
        labels.append(it);
        points.append([point[0] + diff_x, point[1] + diff_y]);

    it += 1;
print(points);
#1 .. 105
#x .. y x = y / 105

px = [];
py = [];

for p in points:
    px.append(p[0]);
    py.append(p[1]);
    #traing_data.append([p[0], p[1]]);
    
#px_norm = preprocessing.normalize(px);
#py_norm = preprocessing.normalize(py);
#plt.plot(traing_data, 'ro');
#plt.plot(px, py, 'bx');
px = preprocessing.normalize([px]);
py = preprocessing.normalize([py]);
plt.plot(px, py, 'ro');

print(px);
print(py);

for i in range(len(px[0])):
    traing_data.append([px[0][i], py[0][i]]);

print("tr: ",traing_data);

#plt.plot(traing_data, 'bo');
#print(traing_data)
plt.show();
```

```python
fig, axs = plt.subplots(ncols=2, nrows=1)
#bins = np.linspace(0, px.max(), 5);
print(px);
axs[0].hist(px[0], bins=50)
axs[1].hist(py[0], bins=50);
```

```python

clf = MLPClassifier().fit(traing_data, labels)


X = np.linspace(px.min(), px.max(), 100);
Y = np.linspace(py.min(), py.max(), 100);

for x in X:
    for y in Y:
        pred = clf.predict([[x, y]])
        if(pred == 1):
            plt.plot(x, y, 'ro');
        if(pred == 2):
            plt.plot(x, y, 'yo');
        if(pred == 3):
            plt.plot(x, y, 'bo');
        if(pred == 4):
            plt.plot(x, y, 'go');
        #plt.plot([x, clf.predict([[x, y]])]);
plt.plot(px, py, 'kx');

plt.show();
```
