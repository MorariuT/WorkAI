```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import random
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
fig = plt.figure()
ax = fig.add_subplot()
```

```python
points = [];
labels = [];
roots = [];
traing_data = []

for i in range(3):
    x = random.randint(0, 100);
    y = random.randint(0, 100);
    print("X: ", x, " y: ", y);
    roots.append([x, y]);
    #points.append([x, y]);

it = 1;
for point in roots:
    for i in range(10):
        diff_x = random.randint(-10, 10);
        diff_y = random.randint(-10, 10);
        labels.append(it);
        points.append([point[0] - diff_x, point[1] - diff_y]);

    it += 1;
print(points);


px = [];
py = [];

for p in points:
    px.append(p[0]);
    py.append(p[1]);
    traing_data.append([p[0], p[1]]);

#print(traing_data)
plt.show();
```

```python
model = make_pipeline(StandardScaler(), SGDClassifier(max_iter=1000, tol=1e-3))
model.fit(traing_data, labels)


for x in range(0, 100, 1):
    for y in range(0, 100, 1):
        pred = model.predict([[x, y]]);
        if(pred == 1):
            plt.plot(x, y, 'bo');
        elif(pred == 2):
            plt.plot(x, y, 'go');
        elif(pred == 3):
            plt.plot(x, y, 'ro');
        #plt.plot(model.predict([[x, y]]));
plt.plot(px, py, 'yx');

plt.show();
```
