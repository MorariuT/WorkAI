```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```python
def f(x, y):
    return (x + y) ** 2;
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

x = np.linspace(0, 100, 100)
y = np.linspace(0, 100, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

ax.plot_surface(X, Y, Z)
```
