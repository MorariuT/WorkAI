```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import random
from mpl_toolkits.mplot3d import Axes3D
```

```python
traing_points = 15;
traing_points_x = [2, 1, 2, 1, 2.1, 2.1, 1.2, 2.1, 6, 5, 5, 5, 6, 5, 1];
traing_points_y = [1, 2, 2, 1, 1.3, 1.1, 1.0, 1, 10, 7, 6, 8, 9, 9, 10];

fig = plt.figure()
ax = fig.add_subplot()

mx = 0;

for i in range(traing_points):
    #x = random.randint(0, 100);
    #y = random.randint(0, 100);
    #traing_points_x.append(x);
    #traing_points_y.append(y);
    mx = max(mx, traing_points_x[i]);
    mx = max(mx, traing_points_y[i]);

ax.plot(traing_points_x, traing_points_y, 'ro');
plt.show();
```

```python
B = 0;
T = 0;

def f(x, B, T):
    return T * x + B;

def Error(b, tetha):
    E = 0;

    for i in range(traing_points):
        x = traing_points_x[i];
        y = traing_points_y[i];
        #print("--> f(x): ", f(x), " y: ", y);
        E += ((f(x, b, tetha) - y)**2)
        
    E /= traing_points;
    return E;
```

```python
pasi = 100000;
e = 0.001;
prec_err = -1;

b_curr = 0;
t_curr = 0;

alpha = 0.0001;

for i in range(pasi):
    y_pred = [];
    for j in range(traing_points):
        x = traing_points_x[j];
        y = traing_points_y[j];

        y_pred.append(f(x, b_curr, t_curr));

    t_deriv = 0;
    b_deriv = 0;

    for j in range(len(y_pred)):
        t_deriv += traing_points_x[j] * (traing_points_y[j] - y_pred[j]);
        b_deriv += traing_points_y[j] - y_pred[j];
    t_deriv *= -1;
    b_deriv *= -1;

    signt = -1;
    if(t_deriv > 0): signt = 1;
    signb = -1;
    if(b_deriv > 0): signb = 1;
    

    t_curr = t_curr - signt * alpha;
    b_curr = b_curr - signb * alpha;

    E = Error(b_curr, t_curr);

    if(i % 1000 == 0):
        print("t: ", t_curr, " b: ", b_curr, " e: ", E, " pas: ", i);


    if(prec_err != -1 and abs(prec_err - E) <= e):
        print("--- Error Not Getting any Better ---");
        break;

    prec_err = E;


print("t: ", t_curr, " b: ", b_curr);
points_function_f = np.arange(0, mx, 1);
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(traing_points_x, traing_points_y, 'ro');
ax.plot(f(points_function_f, b_curr, t_curr), 'b--');
plt.show()
```
